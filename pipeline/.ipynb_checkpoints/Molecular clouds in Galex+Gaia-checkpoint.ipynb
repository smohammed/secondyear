```python
pg = fits.open('../plane_gaiadr2_dust_11_14_18.fits')[1].data
mol = Table.read('../molclouds.txt', format='ascii')
```

```python
d = {}
for cloud in range(len(mol)):
    dcut = np.where((pg['gl'] > mol['llower'][cloud]) & (pg['gl'] < mol['lupper'][cloud]) & (pg['gb'] > mol['blower'][cloud]) & (pg['gb'] < mol['bupper'][cloud]) & (pg['ebv'] < mol['ebv'][cloud]) & (pg['dist'] > mol['dist'][cloud]-(mol['d_err_stat'][cloud]+mol['d_err_syst'][cloud])) & (pg['dist'] < mol['dist'][cloud]+(mol['d_err_stat'][cloud]+mol['d_err_syst'][cloud])))
    d['mol' + str(cloud)] = pg[dcut]
```

```python
print len(np.sort(d['mol2']['ebv']))

fig = plt.figure(figsize=(15,15))
for i in range(1, 15):
    ax = fig.add_subplot(4, 4, i)
    ax.hist(d['mol'+str(i-1)]['ebv'], bins=10, range=[0,1], label=mol['name'][i-1])
    ax.set_xlim([0, 1])
    plt.legend(markerscale=1)

plt.show()
```
